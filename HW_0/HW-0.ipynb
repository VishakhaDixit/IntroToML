```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Read the dataset
df = pd.read_csv("D3.csv")

#Separate the explanatory variables
X1 = df.values[:,0]
X2 = df.values[:,1]
X3 = df.values[:,2]
Y = df.values[:,3]
```

```python
# PROBLEM 1
```

```python
def getMseLoss(Y_known, Y_pred):
    avg_loss = np.square(np.subtract(Y_known, Y_pred)).mean()
    return avg_loss
```

```python
def calGradDescent(X, Y, lrngRate, epoch):
    theta_0 = 0
    theta_1 = 0
    theta = []
    mse = []
    
    for i in range(epoch):
        H = (theta_1 * X) + theta_0    #Compute Predicted output
        J = getMseLoss(Y, H)              #Compute Loss function
        theta_1 = theta_1 - (lrngRate * (-2 * X.dot(Y - H).sum() / (len(X))))
        theta_0 = theta_0 - (lrngRate * (-2 * (Y - H).sum() / (len(X))))
        
        mse.append(J)
        theta.append((theta_0, theta_1))
    
    return theta, mse
```

```python
def getPredictedVal(theta, X):
    return (theta[1]*X + theta[0])
```

```python
# 1. Compute linear model you found for each explanatory variable.
```

```python
it = 100 # Number of iterations
lr = 0.1
print("With LR of ", lr)

thetaX1, mseX1 = calGradDescent(X1, Y, lr, it)
print('Final value of theta for X1 =',thetaX1[99])

thetaX2, mseX2 = calGradDescent(X2, Y, lr, it)
print('Final value of theta for X2 =',thetaX2[99])

thetaX3, mseX3 = calGradDescent(X3, Y, lr, it)
print('Final value of theta for X3 =',thetaX3[99])
```

```python
# 1. Calculate gradient descent for all 3 explainatory variable
def calGradDescent_3(X1, X2, X3, Y, lrngRate, epoch):
    theta_0 = 0
    theta_1 = 0
    theta_2 = 0
    theta_3 = 0
        
    theta = []
    mse = []
    
    for i in range(epoch):
        H = (theta_1 * X1) + (theta_2 * X2) + (theta_3 * X3) + theta_0    #Compute Predicted output
        J = getMseLoss(Y, H)                                              #Compute Loss function
        theta_1 = theta_1 - (lrngRate * (-2 * X1.dot(Y - H).sum() / (len(X1))))
        theta_2 = theta_2 - (lrngRate * (-2 * X2.dot(Y - H).sum() / (len(X2))))
        theta_3 = theta_3 - (lrngRate * (-2 * X3.dot(Y - H).sum() / (len(X3))))
        theta_0 = theta_0 - (lrngRate * (-2 * (Y - H).sum() / (len(X1))))
        
        mse.append(J)
        theta.append((theta_0, theta_1, theta_2, theta_3))
    
    return theta, mse    
```

```python
def getPredictedVal_3(theta, X1, X2, X3):
    return (theta[1]*X1 + theta[2]*X2 + theta[3]*X3 + theta[0])
```

```python
lr = 0.05    # Learning Rate
it = 150     # Epochs
theta_3, mse_3 = calGradDescent_3(X1, X2, X3, Y, lr, it)

lr = 0.01    # Learning Rate
theta_4, mse_4 = calGradDescent_3(X1, X2, X3, Y, lr, it)

lr = 0.1    # Learning Rate
theta_5, mse_5 = calGradDescent_3(X1, X2, X3, Y, lr, it)
```

```python
def getPredictedVal_m(theta, X):
    return np.matmul(theta.T, X.T).T

def calGradDescent_m(X, Y, lrngRate, epoch):
    theta = np.zeros((X.shape[1], 1))
    mse = []
    
    for i in range(epoch):
        H = getPredictedVal_m(theta, X)
        J = getMseLoss(Y, H)
        vectored_grad = 2*np.dot(X.T, H-Y) / (len(Y))
        theta = theta - (lrngRate * vectored_grad)
        mse.append(J)
    
    return theta, mse
```
